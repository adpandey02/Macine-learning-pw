```python
# feature description: 
'''
Fixed acidity: The amount of fixed acids in the wine. These acids contribute to the tartness and 
sourness of the wine. A higher fixed acidity can result in a more acidic taste and may negatively
impact the wine's quality.

Volatile acidity: The amount of volatile acids in the wine. These acids can give the wine a vinegar-like
taste and aroma, which can negatively affect the wine's quality. High levels of volatile acidity are 
usually undesirable in wine.

Citric acid: The amount of citric acid in the wine. Citric acid can contribute to the wine's 
freshness and can enhance its flavor. However, too much citric acid can make the wine too sour or acidic.

Residual sugar: The amount of sugar that remains in the wine after the fermentation process is 
complete. Residual sugar can contribute to the wine's sweetness and can balance out the acidity.
However, too much residual sugar can make the wine taste too sweet.

Chlorides: The amount of salt in the wine. Chlorides can contribute to the wine's flavor and
can help to balance out the sweetness and acidity. However, too much chloride can make the 
wine taste salty.

Free sulfur dioxide: The amount of sulfur dioxide that is added to the wine as a preservative. 
Free sulfur dioxide can prevent the growth of bacteria and preserve the wine's freshness. 
However, too much sulfur dioxide can negatively impact the wine's taste and aroma.

Total sulfur dioxide: The total amount of sulfur dioxide in the wine. Total sulfur dioxide is
an important measure of the wine's quality and stability.

Density: The density of the wine. Density can provide information about the wine's alcohol content 
and sugar concentration.

pH: The level of acidity in the wine. pH can affect the wine's taste, aroma, and color.
Wines with a lower pH are usually more acidic and can taste tart or sour.

Sulphates: The amount of sulphates in the wine. Sulphates can help to preserve the wine and 
prevent the growth of bacteria. However, too much sulphate can negatively impact the wine's
taste and aroma.

Alcohol: The alcohol content of the wine. Alcohol can affect the wine's taste and aroma,
and can contribute to its body and mouthfeel.

Quality: The quality rating of the wine, scored between 0 and 10 by a panel of tasters. 
The quality rating is the dependent variable that the model seeks to predict based on the other features
'''
```

## feature engineering process used

```python
# get a single average score
df['avg_score']=df['math_score']+df['reading_score']+df['writing_score']
```

```python
df['avg_score']=df['avg_score']/3
```

```python
df_backup_copy = df.copy()
```

```python
df.drop(['math_score','reading_score','writing_score'],axis=1, inplace=True)
```

```python
df.head()
```

```python
# label encode gender feature as 1,0 for male and female respectively
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode=LabelEncoder()
```

```python
encoded_gender = encode.fit_transform(df['gender'])
```

```python
df_temp=pd.DataFrame(encoded_gender, columns=['encoded_gender'])
```

```python
df['gender']=encoded_gender
```

```python
df.drop('encoded_gender',axis=1,inplace=True)
```

```python
df  # where now male is 1 , female is 0
```

```python
# one hot encode race_ethnicity feature
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoded=encoder.fit_transform(df[['race_ethnicity']]).toarray()
df_temp = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
df = pd.concat([df.iloc[:,0:1], df_temp, df.iloc[:,2:]], axis=1)
```

```python
df  #with one hot encoded race ethnicity feature into 5 featues based on unique categories
```

```python
# ordinal encode parental_level_of_education and lunch standard
```

```python
df_backup_copy2 = df.copy()     # takeiing backup
```

```python
# merging same looking categories before encoding
```

```python
df.replace(to_replace='some high school',value='high school', inplace=True)
```

```python
df.replace(to_replace='some college',value="bachelor's degree", inplace=True)
```

```python
df.parental_level_of_education.value_counts()
```

```python
#encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder(categories=[['high school',"associate's degree", "bachelor's degree","master's degree"],['free/reduced','standard']])
encoded=encoder.fit_transform(df[['parental_level_of_education','lunch']])
```

```python
df = pd.concat([df.iloc[:,0:6], pd.DataFrame(encoded, columns=encoder.get_feature_names_out()),df.iloc[:,8:10]], 
          axis=1)
```

```python
df # parental_level_of_education	lunch ordinally encoded
```

```python
# label encode test_preparation_course feature 0,1 for none, completed respectively
```

```python
df.test_preparation_course.value_counts()
```

```python
encoder=LabelEncoder()
df['test_preparation_course'] =  encoder.fit_transform(df['test_preparation_course'])
```

```python
# executing corrections on test_preparation_course 0 and 1 flipping
```

```python
df.test_preparation_course =  df.test_preparation_course + 1
df.test_preparation_course.replace(to_replace=2,value=0, inplace=True)
```
